```python
def get_sitemap(url):
    """Scrapes an XML sitemap from the provided URL and returns XML source.

    Args:
        url (string): Fully qualified URL pointing to XML sitemap.

    Returns:
        xml (string): XML source of scraped sitemap.
    """

    response = urllib.request.urlopen(url)
    xml = BeautifulSoup(response, 
                         'lxml-xml', 
                         from_encoding=response.info().get_param('charset'))

    return xml
```

```python
sitemaps = xml.find_all("Statviewtype")
parameters = xml.find_all("Parameters")

data = []
season_id = []
player_id = []
name = []
active = []
age =[]
shoots = []
position = []
games_played = []
team_name = []


#### parameters ###
for parameter in parameters: 
    season_id.append(parameter.findNext("season_id").text)    

### sitemaps ###
    
for sitemap in sitemaps: 
    player_id.append(sitemap.findNext("player_id").text)
    
for sitemap in sitemaps: 
    name.append(sitemap.findNext("name").text)    
    
for sitemap in sitemaps: 
    active.append(sitemap.findNext("active").text)       
    
for sitemap in sitemaps: 
    age.append(sitemap.findNext("age").text)   
    
for sitemap in sitemaps: 
    shoots.append(sitemap.findNext("shoots").text)  
    
for sitemap in sitemaps: 
    position.append(sitemap.findNext("position").text)     

for sitemap in sitemaps: 
    games_played.append(sitemap.findNext("games_played").text)   
    
for sitemap in sitemaps: 
    games_played.append(sitemap.findNext("team_name").text)     


    
data = [player_id, name,active,age,shoots,position,games_played,team_name]
df = pd.DataFrame(data).T
df.columns = ['player_id', 'name','active','age','shoots','position','games_played','team_name']    
```
